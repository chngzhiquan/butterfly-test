To scrape usernames from eBird Trip Report and add to a csv for ebird_arcgis_pipeline

```python
import requests
from bs4 import BeautifulSoup
```

```python
report_url = "https://ebird.org/tripreport/475503"

# 1. Fetch the page
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36'
}
response = requests.get(report_url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# look at the entire page to find the correct class name for participant names
print(soup.prettify()[:2000])
```

```python
# Search for my username to see the HTML structure
my_name = "Zhi Quan Chng"
for tag in soup.find_all(string=lambda text: my_name in text):
    print(f"Tag: {tag.parent.name}")
    print(f"Attributes: {tag.parent.attrs}")
    print(f"Full Tag: {tag.parent}")
```

```python
# 2. Extract Names
participants = []

# Logic: Search for the names in the 'People' or 'Participants' section
# The exact class name often looks like "UserDisplayName" or similar
for name_tag in soup.find_all('a', class_='UserDisplayName'):
    participants.append(name_tag.get_text(strip=True))

# 3. Clean up the list
# Remove duplicates and your own name (the owner)
whitelist = list(set(participants))
print(f"Current Whitelist: {whitelist}")
```
